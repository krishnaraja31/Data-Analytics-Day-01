```python
import pandas as pd
import numpy as np

# File path
file_name = "police.csv"

# Read the CSV file with a specific encoding
try:
    df = pd.read_csv(file_name, encoding='ISO-8859-1')
except UnicodeDecodeError as e:
    print(f"Encoding error: {e}")
    # Handle the error appropriately, e.g., exit or use a different encoding
    exit()

# Data Cleaning and Handling Missing Values

# 1. Check for missing values
print(df.isnull().sum())

# 2. Handling missing values:
#    - For numerical columns, fill with the mean or median
for col in df.select_dtypes(include=np.number):
    if df[col].isnull().any():
        df[col].fillna(df[col].median(), inplace=True) # Use median to avoid outlier influence


#   - For categorical columns, fill with the mode
for col in df.select_dtypes(exclude=np.number):
    if df[col].isnull().any():
        df[col].fillna(df[col].mode()[0], inplace=True)

# 3. Check for duplicates
print(f"\nNumber of duplicate rows: {df.duplicated().sum()}")

# 4. Remove duplicates (if any)
df.drop_duplicates(inplace=True)

# 5. Data Type Conversion (if needed)
# Example: Convert a column to datetime
# if 'date_column' in df.columns:
#     df['date_column'] = pd.to_datetime(df['date_column'])


# Display cleaned data
print("\nCleaned DataFrame:")
print(df.head())
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'police.csv'  # Update with the correct file path
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Set the style for better visuals
sns.set(style="whitegrid")

# Visualization 1: Distribution of Manner of Death
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='manner_of_death', palette='coolwarm',
              order=data['manner_of_death'].value_counts().index)
plt.title('Distribution of Manner of Death', fontsize=16)
plt.xlabel('Manner of Death', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Visualization 2: Count of Incidents by Race
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='race', palette='viridis',
              order=data['race'].value_counts().index)
plt.title('Count of Incidents by Race', fontsize=16)
plt.xlabel('Race', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# Visualization 3: Gender Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='gender', palette='pastel')
plt.title('Gender Distribution', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# Visualization 4: Signs of Mental Illness
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='signs_of_mental_illness', palette='husl')
plt.title('Signs of Mental Illness', fontsize=16)
plt.xlabel('Signs of Mental Illness', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# Visualization 5: Incidents by State (Top 10 States)
top_states = data['state'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_states.index, y=top_states.values, palette='mako')
plt.title('Top 10 States with Most Incidents', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# Visualization 6: Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], kde=True, color='blue', bins=30)
plt.title('Age Distribution of Individuals', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# Visualization 7: Armed Status
plt.figure(figsize=(12, 6))
armed_counts = data['armed'].value_counts().head(10)
sns.barplot(x=armed_counts.index, y=armed_counts.values, palette='flare')
plt.title('Top 10 Armed Statuses', fontsize=16)
plt.xlabel('Armed With', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Visualization 8: Threat Level Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='threat_level', palette='cool')
plt.title('Threat Level Distribution', fontsize=16)
plt.xlabel('Threat Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# Visualization 9: Flee Status
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='flee', palette='autumn')
plt.title('Flee Status Distribution', fontsize=16)
plt.xlabel('Flee Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Visualization 10: Body Camera Usage
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='body_camera', palette='cividis')
plt.title('Body Camera Usage', fontsize=16)
plt.xlabel('Body Camera Used', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
