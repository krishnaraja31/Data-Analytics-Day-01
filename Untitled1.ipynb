```python
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv('kk.csv')
```

```python
# Count missing values for each column
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Percentage of missing values
missing_percentage = data.isnull().mean() * 100
print("\nPercentage of missing values in each column:\n", missing_percentage)

# List of columns with missing values
columns_with_missing = data.columns[data.isnull().any()].tolist()
print("\nColumns with missing values:\n", columns_with_missing)
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('kk.csv')

# Get numeric columns
numeric_cols = data.select_dtypes(include=np.number).columns

# Copy the dataset and fill only numeric columns with their mean
data_mean = data.copy()
data_mean[numeric_cols] = data_mean[numeric_cols].fillna(data_mean[numeric_cols].mean())

print("\nMissing values in numeric columns replaced with Mean.")
#display a preview of updated data
print(data.head())
```

```python
import pandas as pd
import numpy as np

data_median = data.copy()  # Copy the dataset

# Get numeric columns
numeric_cols = data_median.select_dtypes(include=np.number).columns

# Fill missing values in numeric columns with their median
data_median[numeric_cols] = data_median[numeric_cols].fillna(data_median[numeric_cols].median())
print("\nMissing values in numeric columns replaced with Median.")

#display a preview of updated data
print(data.head())

import pandas as pd
import numpy as np

# Assuming data_median is already defined as in your previous code

# Define a function to detect outliers using IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers.index  # Return indices of outliers

# Create an empty dictionary to store outlier information
outliers = {}

# Detect outliers for each numeric column and store in 'outliers' dictionary
for col in numeric_cols:
    outliers[col] = detect_outliers_iqr(data_median, col)

outliers_count = {col: len(outliers[col]) for col in outliers if not outliers[col].empty}
print("\nCount of outliers for each numeric column:\n", outliers_count)
```
